```python
import pandas as pd
from sklearn.model_selection import train_test_split
```

```python
data = pd.read_csv("iris.csv")
data
```

```python
data["Species"].unique()
```

```python
data["Species"] = data["Species"].replace({"Iris-setosa": 1, "Iris-versicolor": 2, "Iris-virginica": 3})
```

```python
data
```

```python
X_train, X_test, y_train, y_test = train_test_split(data[["SepalLengthCm", "SepalWidthCm", "PetalLengthCm", "PetalWidthCm"]], data["Species"], test_size=0.2)
```

```python
X_train
```

```python
from sklearn import linear_model
mymodel = linear_model.LogisticRegression(max_iter=120)
mymodel.fit(X_train,y_train)
```

```python
mymodel.predict(X_test)
```

```python
X_test
```

```python
mymodel.score(X_test, y_test)
```

```python
predicted_output = mymodel.predict(X_test)
predicted_output
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, predicted_output)
cm 
```
